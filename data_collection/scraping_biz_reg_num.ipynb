```python
# 라이브러리 로딩
from os import error
import requests
from bs4 import BeautifulSoup
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
# from selenium.common.exceptions import NoSuchElementException, ElementNotInteractableException

import pandas as pd
import numpy as np
```

```python
# 데이터 로딩
df = pd.read_excel('data/혁신성장_사업자번호조회_포맷.xlsx')
```

```python
# 불필요 문자 리스트
rm_txt_list = ["(주)", "㈜", "(사)", "(유)", "주식회사 ", "주식회사"]

# 불필요한 문자 제거
for i in range(len(df)):
    for rm_txt in rm_txt_list:
        df['기업명'][i] = df['기업명'][i].replace(rm_txt,"")
    print(df['기업명'][i])
```

```python
# 크롬 드라이버 로딩
url = "https://www.creport.co.kr/"
browser = webdriver.Chrome()
browser.maximize_window

# 로그인 계정값
account = "C5048700575"
password = "bax!2021"
full_cols = []

browser.get(url)

# # 아이디 입력
elem = browser.find_element_by_id("mf_wfm_headerIndex_ibx_loginId")
elem.click()
elem.send_keys(account)

# # 패스워드 입력
elem = browser.find_element_by_id("mf_wfm_headerIndex_ibx_passwd")
elem.click()
time.sleep(2)
elem.send_keys(password)
elem.send_keys(Keys.ENTER)
browser.maximize_window()
time.sleep(3)

# elem = browser.find_element_by_xpath('/html/body/div[1]/div[1]/header/a') 
# elem.click()
```

```python
# 사업자등록번호 담을 리스트 선언
list_reg_num = []


for i in range(len(df['기업명'])):
    try:
        # 검색창 지우기
        elem = browser.find_element_by_xpath('//*[@id="mf_wfm_header_ibx_autoSearch"]')
        elem.clear()
        # 기업명 입력
        elem = browser.find_element_by_xpath('//*[@id="mf_wfm_header_ibx_autoSearch"]') 
        elem.send_keys(df['기업명'][i])
        elem.send_keys(Keys.RETURN)
        time.sleep(2)
        # 검색된 기업이 1개인 것만 리스트에 담기 나머지 ""입력
        count = browser.find_element_by_xpath('//*[@id="mf_wfm_content_tab_layout_contents_CRP_CP0201M001_body_tbx_totalCnt_cmpList"]')
        if count.text==str(1):
            elem = browser.find_element_by_xpath('//*[@id="mf_wfm_content_tab_layout_contents_CRP_CP0201M001_body_gen_cmpList_0_bizNo"]')
            list_reg_num.append(elem.text)
        else:
            list_reg_num.append("")
    # 오류 시 pass하고 넘기기
    except:
        list_reg_num.append("")
        pass
    #     time.sleep(2)
        
    
    


```

```python
# 사업자번호 리스트 series로 변환
list_reg_num = pd.Series(list_reg_num)
```

```python
# 기업명 series와 합치기
df_reg_num = pd.concat([df['기업명'], list_reg_num], axis=1)

# 컬럼명 변경
df_reg_num.rename(columns={0:'사업자등록번호'}, inplace=True)
```

```python
# csv로 저장
df_reg_num.to_csv('data/df_reg_num.csv', encoding='cp949')
```
